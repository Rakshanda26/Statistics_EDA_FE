```python
data_copy["Size"]=data_copy["Size"].astype("float")
```

```python
char_to_removes=["+",",","$"]
cols_to_clean=["Installs","Price"]
for item in char_to_removes:
    for col in cols_to_clean:
        data_copy[col]=data_copy[col].str.replace(item,"")
```
